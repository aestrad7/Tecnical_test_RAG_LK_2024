##### Comparison between results of different rags

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Define the data for the models
data = {
    "Model": ["Naive", "Chroma Fusion", "Chroma Multyquery"],
    "Answer Relevancy": [0.9088, 0.8356, 0.8439],
    "Answer Similarity": [0.9161, 0.9187, 0.9293],
    "Answer Correctness": [0.6521, 0.5898, 0.5770],
    "Context Precision": [0.7143, 0.7143, 0.7143],
    "Context Recall": [0.8393, 0.8393, 0.8393],
    "Faithfulness": [0.9551, 0.7821, 0.9231],
    "Context Entity Recall": [0.2269, 0.2269, 0.2388]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Set index
df.set_index("Model", inplace=True)

# Display the table
df_display = df
df_display
```

```python
# Plotting the comparison of the three models

# Transpose dataframe for plotting
df_plot = df.T

# Create a plot
plt.figure(figsize=(10, 6))
for column in df_plot.columns:
    plt.plot(df_plot.index, df_plot[column], marker='o', label=column)
plt.title('Comparison of Model Metrics')
plt.xlabel('Metrics')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend(title="Models")
plt.tight_layout()

# Show the plot
plt.show()
```
